```python
try:
  cross=input("MONOHYBRID:ENTER 1/ DIHYBRID: ENTER 2/")
except Exception as e:
  print(f"An error occurred during cross type input: {e}")
  cross = "invalid"

def monohybrid():
  try:
    a = ""
    while len(a) != 2:
      a=input("Enter character 1 (e.g., 'TT' or 'tt'): ")
      if len(a) != 2:
        print("Invalid input for character 1. Please enter exactly two characters.")
    l1_a=a[0]
    l1_b=a[1]

    b = ""
    while len(b) != 2:
      b=input("Enter character 2 (e.g., 'TT' or 'tt'): ")
      if len(b) != 2:
        print("Invalid input for character 2. Please enter exactly two characters.")
    l2_a=b[0]
    l2_b=b[1]

    print("F1 progeny:",l1_a, l2_a," ",l1_a, l2_b," ",l1_b, l2_a," ",l1_b, l2_b)
    print("F2 progeny:",l1_a, l1_b," ",l1_a, l2_b," ",l2_a, l1_b," ",l2_a, l2_b)
  except Exception as e:
    print(f"An error occurred during monohybrid cross calculation: {e}")

def dihybrid():
  try:
    p1 = ""
    while len(p1) != 4:
      p1 = input("Enter parent 1 genotype (e.g., 'RRYY'): ")
      if len(p1) != 4:
        print("Invalid input for parent 1. Please enter exactly four characters (e.g., 'RRYY').")

    p2 = ""
    while len(p2) != 4:
      p2 = input("Enter parent 2 genotype (e.g., 'rryy'): ")
      if len(p2) != 4:
        print("Invalid input for parent 2. Please enter exactly four characters (e.g., 'rryy').")

    # F1 progeny: Assuming P1 is AABB and P2 is aabb for a simple F1 (AaBb)
    # This is a simplified representation. For a true dihybrid cross, F1 would be uniform (e.g., AaBb)
    f1_genotype = p1[0] + p2[0] + p1[2] + p2[2]
    print("F1 progeny:", f1_genotype, " ", f1_genotype, " ", f1_genotype, " ", f1_genotype)

    # F2 progeny: This part of the F2 calculation is a placeholder and assumes specific allele combinations.
    # For a complete dihybrid F2 Punnett square, there would be 16 possible combinations.
    # This is based on the pattern used in the monohybrid F2 calculation.
    f2_genotype1 = p1[0] + p1[1] + p1[2] + p1[3]
    f2_genotype2 = p1[0] + p2[1] + p1[2] + p2[3]
    f2_genotype3 = p2[0] + p1[1] + p2[2] + p1[3]
    f2_genotype4 = p2[0] + p2[1] + p2[2] + p2[3]

    print("F2 progeny:", f2_genotype1, " ", f2_genotype2, " ", f2_genotype3, " ", f2_genotype4)
  except Exception as e:
    print(f"An error occurred during dihybrid cross calculation: {e}")

if cross == "1":
  monohybrid()
elif cross == "2":
  dihybrid()
else:
  print("Invalid input for cross type.")
```
